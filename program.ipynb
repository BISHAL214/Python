```python
# Write a program in python to count the repeating elemnt in the list
def count_repeating_elements(lst):
    counts = {} 
    for elem in lst:
        if elem in counts:
            counts[elem] += 1
        else:
            counts[elem] = 1 
    repeating_elems = {k:v for k,v in counts.items() if v > 1}
    return repeating_elems

lst = [1, 2, 3, 2, 4, 3, 5, 6, 5]
repeating_elems = count_repeating_elements(lst)
print(repeating_elems)
```

```python
import numpy as np
import matplotlib.pyplot as pit
class plot:
    def mainplot():
        x_axis = np.array([31,41,53,67,87,23,45,46,78,90])
        y_axis = np.array([103,14,34,72,83,34,567,678,456])
        f = pit.figure()
        f.set_figwidth(6)
        f.set_figheight(2)
        pit.plot(x_axis,color="red", marker="o", mfc="purple", mec="black",ms="4")
        pit.plot(y_axis,color="green", marker="o", mfc="yellow", mec="black",ms="4")
        pit.grid(color="blue")
        pit.xlabel("students")
        pit.ylabel("Attendence")
        pit.title("SIT ECE 4th SEMESTER")
        pit.show()
plot.mainplot()
```
